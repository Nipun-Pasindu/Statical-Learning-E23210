### **Q3) Transformer Failures**

1. **Probability Space $(\Omega, \mathfrak{F}, P)$**:
* **Sample Space**: $\Omega = \{0, 1, 2, 3\}$.
* **$\sigma$-algebra**: $\mathfrak{F} = \mathcal{P}(\Omega)$ (the power set of $\Omega$, containing $2^4 = 16$ events).
* **Probability Measure**: $P$ is defined by the given probabilities: $P(\{0\})=0.50$, $P(\{1\})=0.30$, $P(\{2\})=0.15$, $P(\{3\})=0.05$.


2. **Define $X$**: $X$ is a random variable $X: \Omega \to \mathbb{R}$ such that $X(\omega) = \omega$ for all $\omega \in \Omega$.
3. **Probability Mass Function (PMF)**:

$$P(X=x) = \begin{cases} 0.50 & x=0 \\ 0.30 & x=1 \\ 0.15 & x=2 \\ 0.05 & x=3 \\ 0 & \text{otherwise} \end{cases}$$


4. **Expectation $E[X]$**:

$$E[X] = \sum x P(X=x) = 0(0.50) + 1(0.30) + 2(0.15) + 3(0.05) = 0.75$$


5. **Variance $\text{Var}(X)$**:
First, compute $E[X^2] = 0^2(0.50) + 1^2(0.30) + 2^2(0.15) + 3^2(0.05) = 0.30 + 0.60 + 0.45 = 1.35$.

$$\text{Var}(X) = E[X^2] - (E[X])^2 = 1.35 - (0.75)^2 = 1.35 - 0.5625 = 0.7875$$


6. **Probabilities**:
* $P(X \ge 1) = P(X=1) + P(X=2) + P(X=3) = 0.30 + 0.15 + 0.05 = 0.50$.
* $P(X \ge 2) = P(X=2) + P(X=3) = 0.15 + 0.05 = 0.20$.


7. **Interpretation**: $E[X] = 0.75$ means that, on average, there are 0.75 failures reported per day over the long run. $\text{Var}(X) = 0.7875$ measures the daily fluctuation or uncertainty around that average.

### **Q4) Probability of Defective Items**

1. **Probability Space**: $\Omega = [0, 1]$, $\mathfrak{F} = \mathfrak{B}([0, 1])$ (the Borel $\sigma$-algebra), and $P$ is the measure defined by the density function $f(x)$.
2. **Define $X$**: $X$ is the identity random variable $X(\omega) = \omega$ for $\omega \in [0, 1]$.
3. **PDF**: $f(x) = \frac{1}{B(2,18)}x^{1}(1-x)^{17}$ for $0 < x < 1$.
4. **Verify PDF**: The function is non-negative on its support. Integrating over the support:

$$\int_0^1 \frac{x^{2-1}(1-x)^{18-1}}{B(2,18)} dx = \frac{B(2,18)}{B(2,18)} = 1$$


5. **Expectation $E[X]$**: For a Beta distribution, $E[X] = \frac{\alpha}{\alpha + \beta}$.

$$E[X] = \frac{2}{2 + 18} = \frac{2}{20} = 0.10$$


6. **Variance $\text{Var}(X)$**: $\text{Var}(X) = \frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$.

$$\text{Var}(X) = \frac{2 \cdot 18}{(20)^2(21)} = \frac{36}{8400} = \frac{3}{700} \approx 0.004286$$


7. **$P(X \le 0.10)$**: $P(X \le 0.10) = \int_0^{0.10} \frac{x(1-x)^{17}}{B(2,18)} dx$.
8. **$P(0.05 \le X \le 0.15)$**: $P(0.05 \le X \le 0.15) = \int_{0.05}^{0.15} \frac{x(1-x)^{17}}{B(2,18)} dx$.
9. **Reasonability**: The Beta distribution is defined on $[0, 1]$, making it mathematically ideal for modeling proportions or probabilities.
10. **Interpretation**: $E[X]=10\%$ suggests the expected defect rate. The small variance indicates that the daily defect rate is consistently close to this average.

### **Q5) Expected Project Completion Time**

1. **Support**: $[4, 10]$.
2. **PDF**:

$$f_X(x) = \begin{cases} \frac{2(x-4)}{(10-4)(6-4)} = \frac{x-4}{6} & 4 \le x < 6 \\ \frac{2(10-x)}{(10-4)(10-6)} = \frac{10-x}{12} & 6 \le x \le 10 \end{cases}$$


3. **Verify Integrates to 1**: The area of the triangle is $\frac{1}{2} \times \text{base} \times \text{height}$. Base $= 10-4 = 6$. Height $= f_X(6) = \frac{6-4}{6} = \frac{1}{3}$. Area $= \frac{1}{2} \cdot 6 \cdot \frac{1}{3} = 1$.
4. **Expectation $E[X]$**: $E[X] = \frac{a+b+c}{3} = \frac{4+10+6}{3} = \frac{20}{3} \approx 6.67$ hours.
5. **Variance $\text{Var}(X)$**: $\text{Var}(X) = \frac{a^2+b^2+c^2-ab-ac-bc}{18} = \frac{16+100+36-40-24-60}{18} = \frac{28}{18} = \frac{14}{9} \approx 1.56$.
6. **Interpret $a, b, c$**: $a=4$ is the optimistic (shortest) time, $b=10$ is the pessimistic (longest) time, and $c=6$ is the most likely time.
7. **Reasonability**: It only requires three intuitive estimates and can model skewed data where the most likely time is not the average.
8. **$P(X \le 5)$**: Using the first piece of the PDF: $\int_4^5 \frac{x-4}{6} dx = [\frac{(x-4)^2}{12}]_4^5 = \frac{1}{12} \approx 0.0833$.
9. **$P(5 \le X \le 8)$**: $P(X \le 8) - P(X \le 5) = (1 - \frac{(10-8)^2}{(10-4)(10-6)}) - \frac{1}{12} = (1 - \frac{4}{24}) - \frac{1}{12} = \frac{5}{6} - \frac{1}{12} = 0.75$.
10. **Application**: Used for project scheduling (CPM/PERT) to estimate duration and perform risk analysis.

### **Q6) Expected Project Completion Time Using PERT**

1. **Shape Parameters**:

$$\alpha = 1 + 4\left(\frac{11-8}{20-8}\right) = 1 + 4\left(\frac{3}{12}\right) = 2$$


$$\beta = 1 + 4\left(\frac{20-11}{20-8}\right) = 1 + 4\left(\frac{9}{12}\right) = 1 + 3 = 4$$


2. **Support**: $[8, 20]$.
3. **PDF**: $f_X(x) = \frac{(x-8)^{2-1}(20-x)^{4-1}}{B(2,4)(20-8)^{2+4-1}} = \frac{(x-8)(20-x)^3}{B(2,4) \cdot 12^5}$ for $8 \le x \le 20$.
4. **Probability Space**: $\Omega = [8, 20]$, $\mathfrak{F} = \mathfrak{B}([8, 20])$, and $P$ defined by the PERT density.
5. **Define $X$**: $X$ is the random variable mapping the sample space to the range $[8, 20]$.
6. **Show $E[X]$**: The PERT distribution is a Beta distribution on $[0,1]$ scaled to $[a,b]$. $E[X] = a + (b-a)\frac{\alpha}{\alpha+\beta}$.
Since $\alpha+\beta = (1+4\frac{c-a}{b-a}) + (1+4\frac{b-c}{b-a}) = 2 + 4 = 6$,
$E[X] = a + (b-a)\frac{1+4\frac{c-a}{b-a}}{6} = \frac{6a + (b-a) + 4(c-a)}{6} = \frac{6a + b - a + 4c - 4a}{6} = \frac{a+4c+b}{6}$.
7. **Variance**: $E[X] = \frac{8 + 4(11) + 20}{6} = \frac{72}{6} = 12$.

$$\text{Var}(X) = \frac{(12-8)(20-12)}{7} = \frac{4 \cdot 8}{7} = \frac{32}{7} \approx 4.571$$


8. **Standard Deviation**: $\sigma = \sqrt{32/7} \approx 2.138$ days.
9. **Comparison**: PERT is smoother (bell-shaped) than the Triangular distribution and places less emphasis on the extreme values ($a$ and $b$), typically leading to a more realistic estimation in software development.
10. **Interpretation**: $a$ is optimistic, $b$ is pessimistic, $c$ is the mode, $E[X]$ is the "weighted" average task duration, and $\sigma$ represents the risk/variability of the task.

### **Q7) If the PERT Expectation is to Match Our Intuition**

1. **Value of $b$**: We are given $a=5$, $c=11$, and we want $E[X] = c$.
Using $E[X] = \frac{a+4c+b}{6} = c$, we get:

$$a+4c+b = 6c \implies b = 2c - a$$


$$b = 2(11) - 5 = 17 \text{ days}$$


2. **Variance**: When $b = 2c - a$, the distribution becomes symmetric ($\alpha = \beta = 3$). The variance is calculated using the range $b-a$:

$$\text{Var}(X) = \frac{(c-a)(b-c)}{7} = \frac{(11-5)(17-11)}{7} = \frac{6 \cdot 6}{7} = \frac{36}{7} \approx 5.14$$
