## Q1)
A disk has radius $R=10$. Event $A$ is the point lying within 1 unit of the boundary (the annulus between $r=9$ and $r=10$).

1.  **Model 1 (Uniform Area):**
    * $\text{Total Area} = \pi(10)^2 = 100\pi$.
    * $\text{Area of } A = \text{Total Area} - \pi(9)^2 = 100\pi - 81\pi = 19\pi$.
    * $P(A) = \frac{19\pi}{100\pi} = \mathbf{0.19}$.
2.  **Model 2 (Uniform Distance):**
    * Since $r \sim \text{Uniform}(0, 10)$, the probability is based on the interval length.
    * $P(A) = P(9 \leq r \leq 10) = \frac{10-9}{10-0} = \mathbf{0.10}$.
3.  **Difference:** Model 1 weights probability by physical area (more space near the edge), while Model 2 weights every radial step equally, effectively making the center "denser" than the edge compared to Model 1.

## Q2)
The goal is to find the probability that a random chord is longer than the side of an inscribed equilateral triangle ($L = R\sqrt{3}$).

* **Random Endpoints:** Fix one point and pick another. The chord is longer if the second point is in the middle $1/3$ arc. $P = \mathbf{1/3}$.
* **Random Radial Distance:** Pick a radius and a point on it. The chord is longer if the point is within $R/2$ of the center. $P = \mathbf{1/2}$.
* **Random Midpoint:** Pick any point in the circle. The chord is longer if the point falls in a concentric circle with radius $R/2$ (area is $1/4$ of the total). $P = \mathbf{1/4}$.

## Q8)
* **Assumptions:** Each child is equally likely to be B or G ($1/2$ each), independently.
* **Sample Space ($\Omega$):** $\{BB, BG, GB, GG\}$, each with probability $1/4$.
* **Condition:** "At least one is a girl" reduces the space to $\{BG, GB, GG\}$.
* **Calculation:**
    $$P(GG \mid \text{at least one G}) = \frac{P(GG)}{P(BG) + P(GB) + P(GG)} = \frac{1/4}{3/4} = \mathbf{1/3}$$

## Q9)
Items are $G_1, G_2$ (Good) or $D_1, D_2$ (Defective).

1.  **$\sigma(X)$:** $X$ only knows if an item is Good ($X=0$) or Defective ($X=1$). $\sigma(X) = \{\emptyset, \Omega, \{G_1, G_2\}, \{D_1, D_2\}\}$.
2.  **$\sigma(Y)$:** $Y$ knows quality AND machine. $\sigma(Y) = \mathfrak{P}(\Omega)$ (all 16 possible subsets).
3.  **Marginal $X$:** $P(X=0) = 0.5+0.2 = 0.7$; $P(X=1) = 0.1+0.2 = 0.3$.
4.  **Conditional Probabilities:**
    * $P(X=1 \mid Y=3) = \mathbf{1}$ (If we know it's $D_1$, it's definitely defective).
    * $P(Y=3 \mid X=1) = \frac{P(D_1)}{P(D_1)+P(D_2)} = \frac{0.1}{0.3} = \mathbf{1/3}$.
5.  **Entropy:** $H(X)$ measures uncertainty in quality; $H(Y)$ measures uncertainty in both quality and machine. Since $\sigma(X) \subseteq \sigma(Y)$, $H(Y) \geq H(X)$.

## Q10)
$T \sim N(0,1)$ and $P \sim N(0,4)$ are independent deviations.

1.  **$\sigma(X)$ vs $\sigma(Y)$:** $\sigma(X)$ contains events involving only temperature (vertical strips in the $T-P$ plane). $\sigma(Y)$ contains all Borel sets in $\mathbb{R}^2$.
2.  **Marginal $X$:** $f_X(t) = \frac{1}{\sqrt{2\pi}}e^{-t^2/2}$.
3.  **Probabilities:**
    * $P(X \leq 0) = \mathbf{0.5}$ (due to symmetry of Normal distribution).
    * $P(Y \in [-1, 1] \times [-2, 2]) = P(-1 \leq T \leq 1) \cdot P(-2 \leq P \leq 2)$.
4.  **Differential Entropy:**
    * $h(X) = \frac{1}{2}\ln(2\pi e \sigma_T^2) = \frac{1}{2}\ln(2\pi e)$.
    * $h(Y) = h(T) + h(P) = \frac{1}{2}\ln(2\pi e \cdot 1) + \frac{1}{2}\ln(2\pi e \cdot 4)$.
5.  **$h(X \mid Y)$:** In the continuous case, if $Y$ reveals $X$ exactly, $h(X \mid Y) \to -\infty$. Unlike discrete entropy, continuous entropy can be negative when a value is known with infinite precision.
